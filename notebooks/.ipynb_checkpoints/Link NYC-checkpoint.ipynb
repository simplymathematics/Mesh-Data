```python
installation = 12900
fiber_connection = 15000
hardware = 5000
maintenance = 1400
power = 2400
bandwidth = 8400
ad_revenue = 60000
```

```python
def kiosk_install_cost(k):  
    return -installation*k - fiber_connection*k- hardware*k 
def kiosk_ongoing_cost(year, k, ad_revenue):
    return -maintenance*k*year - power*k*year - bandwidth*k*year + ad_revenue*k*year
def kiosk_yearly(year, k, ad_revenue):
    return kiosk_install_cost(k)+kiosk_ongoing_cost(year, k, ad_revenue)
```

```python
print(kiosk_install_cost(1))
print(kiosk_ongoing_cost(1,1,60000))
print(kiosk_yearly(1,1,60000))
print(kiosk_yearly(10,1,0))
```

```python
def kiosk_projection(ad_revenue, number=10000):
    grx = plt.plot()
    this_list = []
    for k in range(number):
        this_list.append(kiosk_yearly(10,k,ad_revenue))
    return(grx)
```

```python
this_list = np.linspace(0,100000,10)

for revenue in this_list:
    kiosk_projection(revenue, number=10000)
    grx += plt.plot(this_list)
    
kiosk_projection(30000)
```
